```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("headbrain.csv")
```

```python
X_train = data[:200]
X_test = data[200:]
X = X_train['Head Size(cm^3)'].values
Y = X_train['Brain Weight(grams)'].values
mean_X = np.mean(X)
mean_Y = np.mean(Y)
m = len(X)
print (m)
```

```python
numer = 0
denom = 0
for i in range(m):
    numer += ((X[i] - mean_X) * (Y[i] - mean_Y))
    #numer += (X[i] - mean_X) * (Y[i] - mean_Y)
    denom += (X[i] - mean_X) ** 2
b1 = numer/denom
b0 = mean_Y - (b1 * mean_X)
print(b1,b0)
```

```python
max_x = np.max(X)
min_x = np.min(X)
```

```python
x = np.linspace(min_x,max_x)
y = b0 + b1*x
plt.plot(x,y,color = 'YELLOW',label = 'Regression LIne')
plt.scatter(X,Y,color = "Green",label = 'Scatter plot')
plt.xlabel('Head Size in (cm^3)')
plt.ylabel('Brain weight in(grams)')
plt.legend()
plt.show()
```

```python
mse = 0
for i in range(m):
    y_pred = b0 + b1 * X[i]
    mse += (Y[i] - y_pred) ** 2       

print("Mean sqaure error of brain weight of train data",mse)
rmse = np.sqrt(mse/m)
print("Root Mean sqaure error is",rmse)
```

```python
ss_t = 0
ss_r = 0
for i in range(m):
    y_pred = b0 + b1 * X[i]
    ss_t = (Y[i] - mean_Y) ** 2
    ss_r = (Y[i] - y_pred) ** 2
r2 = 1 - (ss_r/ss_t)
print("R^2 score for training data is",r2)
```
